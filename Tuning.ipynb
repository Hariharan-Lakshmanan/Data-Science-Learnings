```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
import pandas as pd
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['flower'] = iris.target
df['flower'] = df['flower'].apply(lambda x: iris.target_names[x])
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear', C=30, gamma='auto')
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
df = pd.DataFrame(data=digits.data, columns=digits.feature_names)
df['digit'] = digits.target
df['digit'] = df['digit'].apply(lambda x: digits.target_names[x])
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2)
```

Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(SVC(), {'C': [1, 10, 100, 1000], 'gamma': [0.001, 0.0001], 'kernel' : ['rbf', 'linear']}, cv=5)
clf.fit(digits.data, digits.target)
#clf.cv_results_
df = pd.DataFrame(clf.cv_results_)
df.head()
```

```python
df[['param_C', 'param_gamma', 'param_kernel', 'mean_test_score']].sort_values(by='mean_test_score', ascending=False).head(10)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

Randomied Search  CV

```python
from sklearn.model_selection import RandomizedSearchCV
clf = RandomizedSearchCV(SVC(gamma='auto'), {'C': [1, 10, 100, 1000], 'kernel' : ['rbf', 'linear']}, cv=5, n_iter=7)
clf.fit(digits.data, digits.target)
df = pd.DataFrame(clf.cv_results_)[['param_C', 'param_kernel', 'mean_test_score']]
df
```

Different models with different hyper parameters

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB

model_params = {
    'svm': {
        'model': SVC(gamma='auto'),
        'params': {
            'C': [1, 10, 20],
            'kernel': ['rbf', 'linear'],
            'degree': [3, 5],  # for poly kernel (if used)
        }
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [10, 50, 100],
            'max_depth': [None, 5, 10],
            'min_samples_split': [2, 5],
        }
    },
    'logistic_regression': {
        'model': LogisticRegression(solver='liblinear', multi_class='auto'),
        'params': {
            'C': [0.1, 1, 10],
            'penalty': ['l1', 'l2']
        }
    },
    'decision_tree': {
        'model': DecisionTreeClassifier(),
        'params': {
            'max_depth': [None, 5, 10, 20],
            'min_samples_split': [2, 5, 10],
            'criterion': ['gini', 'entropy']
        }
    },
    'gaussian_nb': {
        'model': GaussianNB(),
        'params': {
            # GaussianNB usually has no tunable hyperparameters, but you can test priors
            'var_smoothing': [1e-9, 1e-8, 1e-7]
        }
    },
    'multinomial_nb': {
        'model': MultinomialNB(),
        'params': {
            'alpha': [0.5, 1.0, 1.5],
            'fit_prior': [True, False]
        }
    }
}
```

```python
import warnings
warnings.filterwarnings('ignore')

scores = []
for model_name, mp in model_params.items():
    clf = GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    clf.fit(digits.data, digits.target)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })

df = pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
df.sort_values(by='best_score', ascending=False)
```
