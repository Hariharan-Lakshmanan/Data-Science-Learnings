```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
iris.target_names
```

```python
df['target'] = iris.target
df.head()
```

```python
df[df.target == 1].head()
```

```python
df['flower_name'] = df.target.apply(lambda x : iris.target_names[x])
df.head()
```

```python
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
```

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color= 'red', marker= 'o', label= 'setosa')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color= 'blue', marker= 's', label= 'versicolor')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
```

```python
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color= 'red', marker= 'o', label= 'setosa')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color= 'blue', marker= 's', label= 'versicolor')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'], color= 'green', marker= 'x', label= 'virginica')
plt.xlabel('petal Length')
plt.ylabel('petal Width')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target', 'flower_name'], axis= 1)
X.head()
```

```python
y = df['target']
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2)
```

```python
len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict([[4.7, 3.2, 1.3, 0.2]])
```

Excercise - Digits Dataset

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
df = pd.DataFrame(digits.data, columns=digits.feature_names)
df.head()
```

```python
df['target'] = digits.target
df.head()
```

```python
df.target.value_counts()
```

```python
X = df.drop(['target'], axis = 1)
X.head()
```

```python
y = df['target']
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2)
```

```python
model2 = SVC(kernel= 'rbf', gamma='scale')
```

```python
model2.fit(X_train, y_train)
```

```python
model2.score(X_test, y_test)
```

```python
model2.predict([X_test.iloc[0]])
```
